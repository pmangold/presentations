```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,1000)
y1 = np.exp(-np.abs(x))
y2 = np.exp(-np.abs(x+0.5))
```

```python
plt.plot(x, y1, lw=3, c="red")
plt.plot(x, y2, lw=3, c="blue")
plt.axis("off")
plt.savefig("distrib.pdf", bbox_inches="tight")
```
